```python
import numpy as np
import pandas as pd
from numpy.random import randn

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot('total_bill','tip',tips)
```

```python
sns.lmplot('total_bill','tip',tips,markers='^',
           scatter_kws={'color':'indianred'},
          line_kws={'linewidth':3,'color':'blue'})
```

```python
sns.lmplot('total_bill','tip',tips,markers='^',order=4,
           scatter_kws={'color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
```

```python
sns.lmplot('total_bill','tip',tips,fit_reg=False)
```

```python
tips['tip_pect']=100*(tips['tip']/tips['total_bill'])
```

```python
tips.head()
```

```python
sns.lmplot('size','tip_pect',tips)
```

```python
sns.lmplot('size','tip_pect',tips,x_jitter=.1)
```

```python
sns.lmplot('size','tip_pect',tips,x_estimator=np.mean)
```

```python
sns.lmplot('total_bill','tip_pect',tips,hue='day',markers=['x','o','^','v'])
plt.ylim(0,50)
```

```python
sns.lmplot('total_bill','tip_pect',tips,lowess=True,line_kws={'color':'black'})
```

```python
sns.regplot('total_bill','tip_pect',tips)
```

```python
fig, (axis1,axis2) = plt.subplots(1,2,sharey=False)

sns.regplot('total_bill','tip_pect',tips,ax=axis1)
sns.violinplot(x='size',y='tip_pect',palette='Reds_r',data=tips,ax=axis2)
```
